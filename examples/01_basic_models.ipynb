```python
from pyqueueing import MM1, MMC, MM1K, MMcK, MG1, ErlangB, ErlangC
```

## M/M/1 — Single Server Queue

The simplest queueing model: one server, Poisson arrivals, exponential service.

```python
q = MM1(arrival_rate=2.0, service_rate=3.0)

print(f"Utilization ρ:       {q.utilization():.4f}")
print(f"Mean queue length Lq: {q.mean_queue_length():.4f}")
print(f"Mean system size L:   {q.mean_system_size():.4f}")
print(f"Mean wait time Wq:    {q.mean_wait():.4f}")
print(f"Mean system time W:   {q.mean_system_time():.4f}")
print()
# Verify Little's Law: L = λW
lam = 2.0
print(f"Little's Law check: L = {q.mean_system_size():.4f}, λW = {lam * q.mean_system_time():.4f}")
```

## M/M/c — Multi-Server Queue

```python
q = MMC(arrival_rate=10, service_rate=3, servers=4)

print(f"Utilization ρ:        {q.utilization():.4f}")
print(f"P(wait) [Erlang C]:   {q.prob_wait():.4f}")
print(f"Mean wait time Wq:    {q.mean_wait():.4f}")
print(f"Mean queue length Lq: {q.mean_queue_length():.4f}")
```

## M/M/1/K — Finite Capacity Queue

```python
q = MM1K(arrival_rate=2.0, service_rate=3.0, capacity=10)

print(f"Blocking probability: {q.prob_block():.6f}")
print(f"Effective arrival rate: {q.effective_arrival_rate():.4f}")
print(f"Mean system size L: {q.mean_system_size():.4f}")
print()
# ρ > 1 is OK for finite capacity
q2 = MM1K(arrival_rate=5.0, service_rate=3.0, capacity=10)
print(f"ρ>1 case — Blocking: {q2.prob_block():.4f}, L: {q2.mean_system_size():.4f}")
```

## M/G/1 — General Service Distribution

Compare M/M/1 (CV=1), M/D/1 (CV=0), and high-variability (CV=2).

```python
lam, mu = 2.0, 3.0

for cv in [0.0, 0.5, 1.0, 1.5, 2.0]:
    q = MG1(arrival_rate=lam, service_rate=mu, service_cv=cv)
    print(f"CV={cv:.1f}  Lq={q.mean_queue_length():.4f}  Wq={q.mean_wait():.4f}")
```
